```python
def fill_mis(df):
    total = df.isnull().sum().sort_values(ascending=False)
    mis_data=pd.DataFrame(total).reset_index()
    mis_data.columns=['Features','Total']
    x=np.array(mis_data['Features'][mis_data['Total']>0])
    for i in range(0,len(x)):
        mean = df[x[i]].mean()
        median = df[x[i]].median()
        std = df[x[i]].std()
        if std >10:
            df[x[i]]=df[x[i]].fillna(median)
        else:
            df[x[i]]=df[x[i]].fillna(mean)
```
