```python
import pandas as pd
import numpy as np

data = pd.read_csv('data_for_model.csv')
data.head()
```

```python
data['tag']=data['tag'].astype('object')
data['greatstone_rating']=data['greatstone_rating'].astype('category')
data.dtypes
```

```python
data['below_b_rating'] =data['below_b_rating'].apply(lambda x: x+2.0)
data['b_rating'] =data['b_rating'].apply(lambda x: x+2.0)

data['return_ytd'] =data['return_ytd'].apply(lambda x: x+26.0)
data['2018_return_category'] =data['2018_return_category'].apply(lambda x: x+29.0)

data['1_year_return_fund'] =data['1_year_return_fund'].apply(lambda x: x+38.0)
data['fund_return_3years'] =data['fund_return_3years'].apply(lambda x: x+29.0)

data['2018_return_fund'] =data['2018_return_fund'].apply(lambda x: x+61.0)
data['1_month_fund_return'] =data['1_month_fund_return'].apply(lambda x: x+15.0)

data['10yrs_treynor_ratio_category'] =data['10yrs_treynor_ratio_category'].apply(lambda x: x+2.0)
data['10_years_return_fund'] =data['10_years_return_fund'].apply(lambda x: x+36.0)

data['10_years_alpha_category'] =data['10_years_alpha_category'].apply(lambda x: x+2.0)
data['3yrs_treynor_ratio_fund'] =data['3yrs_treynor_ratio_fund'].apply(lambda x: x+4147.0)

data['3_years_alpha_fund'] =data['3_years_alpha_fund'].apply(lambda x: x+37.0)
data['3yrs_sharpe_ratio_fund'] =data['3yrs_sharpe_ratio_fund'].apply(lambda x: x+6.0)

data['3_years_alpha_category'] =data['3_years_alpha_category'].apply(lambda x: x+2.0)
data['5_years_return_mean_annual_category'] =data['5_years_return_mean_annual_category'].apply(lambda x: x+2.0)

data['5_years_return_fund'] =data['5_years_return_fund'].apply(lambda x: x+25.0)
data['5_years_alpha_category'] =data['5_years_alpha_category'].apply(lambda x:x+2.0)
```

```python
data_log = pd.DataFrame()
df_x =data[data.select_dtypes(include=['int32','int64','float64','float32']).columns]
x=df_x.columns
for i in range(0,len(x)):
    data_log['log_'+x[i]] = np.log(data[x[i]])
```

```python
data_log.head()
```

```python
data_log['greatstone_rating']=data['greatstone_rating']
```

```python
data_log.describe(include='all')
```

```python
X = data_log.drop(columns=['greatstone_rating'])
y = data_log['greatstone_rating']
print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
KNN = KNeighborsClassifier(n_neighbors= 5 , weights = 'distance' )
```

```python
KNN.fit(X_train, y_train)
```

```python
predicted_val_labels = KNN.predict(X_test)
KNN.score(X_test, y_test)
```

```python
from sklearn import metrics

print(metrics.confusion_matrix(y_test, predicted_val_labels))
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

myList = list(range(1,50))

cv_scores = []
k_neighbors = []

for k in myList:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X_train, y_train, cv=10, scoring='accuracy')
    cv_scores.append(scores.mean())
    k_neighbors.append(k)
    
best_k = myList[cv_scores.index(max(cv_scores))]
print ("The optimal number of neighbors is %d" % best_k)
```

```python
def fill_mis(df):
    total = df.isnull().sum().sort_values(ascending=False)
    mis_data=pd.DataFrame(total).reset_index()
    mis_data.columns=['Features','Total']
    x=np.array(mis_data['Features'][mis_data['Total']>0])
    for i in range(0,len(x)):
        mean = df[x[i]].mean()
        median = df[x[i]].median()
        std = df[x[i]].std()
        if std >10:
            df[x[i]]=df[x[i]].fillna(median)
        else:
            df[x[i]]=df[x[i]].fillna(mean)
```
